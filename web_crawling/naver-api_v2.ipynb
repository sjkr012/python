```python
# 네이버 Developers 오픈 API 호출 예제 - python 버전을 참고하였습니다
# https://developers.naver.com/docs/search/blog/

# 잔재미코딩 "Open API를 이용한 웹 크롤링"을 참고하였습니다
# https://www.fun-coding.org/crawl_basic3.html

# 문제 시 바로 삭제하겠습니다

# 뉴스 open API 사용법 with requests
import os
import sys
import urllib.request  
import json
import requests
import pandas as pd
import pickle

def save_to_pickle(df,search_title):
    pickle.dump(df, open('./data/{}.pkl'.format(search_title),'wb'))
    print('저장완료')

def remove_tags(str1):
    # 그냥 간단하게 <b> </b> &quot; 제거하기
    if "<b>" in str1:
        str1 = str1.replace("<b>","")
        str1 = str1.replace("</b>","")
    if "&quot;" in str1:
        str1 = str1.replace("&quot;","")
    
    return str1


def data_to_df(data):
    df = pd.DataFrame(columns=['title', 'description'])

    for i in range(len(data['items'])):
        
        titl = remove_tags(data['items'][i]['title'])
        desc = remove_tags(data['items'][i]['description'])

        df.loc[len(df)] = {
            'title':titl,
            'description':desc
        }
    save_to_pickle(df, 'api_use')

# 아이디, 비밀번호 설정
client_id = "YOUR_CLIENT_ID"  #YOUR_CLIENT_ID
client_secret = "YOUR_CLIENT_SECRET" #YOUR_CLIENT_SECRET

# query,url 및 header paramaters 설정
query = '물병'
naver_url = 'https://openapi.naver.com/v1/search/news.json?query=' + query
header_params = {"X-Naver-Client-Id":client_id, "X-Naver-Client-Secret":client_secret}

# header 변경 시에만 headers=header_params가 필요함
response = requests.get(naver_url, headers = header_params)

# Requests 안의 text를 사용하여 전체 api 호출 가능
#print(response.text)

# SUCCESS 했을 시
if(response.status_code == 200):
    data = response.json()
    data_to_df(data)
    print(data['items'][1]['title'])
    print(data['items'][1]['description'])
else:
    print("Error Code:" + response.status_code)
```
